```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('car.jpg', cv2.IMREAD_GRAYSCALE)  # read as grayscale, default filter reads as colored image but without alpha
""" 
Other options:  
IMREAD_GRAYSCALE = 0
IMREAD_COLOR = 1, 
IMREAD_UNCHANGED = -1
"""

```

## Show the image using CV

```python
title='image'
cv2.imshow(title, img)  # loads the image
cv2.waitKey(0)  # waits for any key
cv2.destroyAllWindows()  # once a key is pressed, destroys the window
```

## Write the gray image to a file using CV

```python
cv2.imwrite('car_gray.png', img)
```

## Show the image using Matplotlib

```python
plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.plot([50, 1000], [80, 1000], 'c', linewidth=5)  # it's possible to plot directly on the image
plt.show()
```
