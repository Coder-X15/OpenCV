```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# load two images of identical size
img1 = cv2.imread('3D-Matplotlib.png')
img2 = cv2.imread('mainsvmimage.png')
```

## Basic addition of images

```python
add = img1 + img2  # add the two images to one
plt.imshow(add, interpolation='bicubic')
```

## CV2 addition of images

```python
add = cv2.add(img1, img2)  # cv2 summed up the RGB values of both pictures
plt.imshow(add, interpolation='bicubic')
```

## Weighted addition of images

```python
weighted = cv2.addWeighted(img1, 0.6, img2, 0.4, 0)  # give the 1st image a weight of 0.6 and 0.4 for the 2nd
plt.imshow(weighted, interpolation='bicubic')
```

## Masking
convert anything below 220 to black
convert anything above 220 to white
then, invert the colors

we use the mask to get only a part of the original picture

```python
img2 = cv2.imread('mainlogo.png')

rows, cols, channels = img2.shape  # copy the image dimensions
roi = img1[0:rows, 0:cols]

img2gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)


min_val = 220
max_val = 255
ret, mask = cv2.threshold(img2gray, 220, 255, cv2.THRESH_BINARY_INV)  # conver


mask_inv = cv2.bitwise_not(mask)  # use bitwise not to get everything which is not on the mask

# set the masks
img1_bg = cv2.bitwise_and(roi, roi, mask=mask_inv)
img2_fg = cv2.bitwise_and(img2, img2, mask=mask)

dst = cv2.add(img1_bg, img2_fg)
img1[0:rows, 0:cols] = dst

plt.imshow(weighted, interpolation='bicubic')  # doesn't work good
cv2.imshow('gray', img2gray)
cv2.imshow('thresh', mask)
cv2.imshow('img1_bg', img1_bg)
cv2.imshow('img2_fg', img2_fg)
cv2.imshow('img1_bg + img2_fg', dst)
cv2.imshow('res',img1)
```
